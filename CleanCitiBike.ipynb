```python
import pandas as pd
import numpy as np
import datetime
from dateutil.relativedelta import relativedelta
from datetime import date
```

```python
#Read the csv file
citiBike_df = pd.read_csv('Resources/citiBike.csv')
```

```python
#sample the first five records
citiBike_df.head()
```

```python
#delete the last column not part of the raw data
del citiBike_df['Unnamed: 13']
```

```python
#display count to see if there is any na value. Indeed there are NAN values at headers: end_station_name,and end_station_id
citiBike_df.count()
```

```python
#drop all na values
citiBike_df = citiBike_df.dropna(how = 'any')
```

```python
#verify the all na values are dropped. Indeed, all na values are dropped
citiBike_df.count()
```

```python
#convert the started_at column values to datetime64 and create a new column: start_time
citiBike_df['start_datetime'] = pd.to_datetime(citiBike_df['started_at'])
```

```python
#convert the ended_at column values to datetime64 and create a new column: end_time
citiBike_df['end_datetime'] = pd.to_datetime(citiBike_df['ended_at'])
```

```python
# verify the conversion
citiBike_df.head()
```

```python
# Calculate the total ride time for each rider and create a new coulmn name: total_ride_time (the values are in seconds)

citiBike_df['total_ride_time'] =citiBike_df['end_datetime'] - citiBike_df['start_datetime']
citiBike_df['tripduration (minutes)']=citiBike_df['total_ride_time']/np.timedelta64(1,'m')
```

```python
#verify the new dataframe
citiBike_df.tail(100)
```

```python
#Recreate new columns and rename the dataframe as cleaned_citibike_df

cleaned_citibike_df = citiBike_df[['tripduration (minutes)','ride_id','rideable_type', 'start_datetime','end_datetime','start_station_name','start_station_id', 'end_station_name','end_station_id', 'start_lat','start_lng','end_lat', 'end_lng', 'member_casual']]
```

```python
#verify the dataframe
cleaned_citibike_df.tail()
```

```python
#verify all the data types before saving
cleaned_citibike_df.dtypes
```

```python
#save the cleaned_citibike_df for tableau visualization

cleaned_citibike_df.to_csv("Resources/clean_citibike.csv", index=False)
```
